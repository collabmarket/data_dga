```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display, HTML
import ddaa

%matplotlib inline
```

```python
# Region a analizar
region = 'V'
```

```python
# Genera DataFrame con ddaa
meta = ddaa.open_meta()
df = ddaa.open_ddaa(meta['path'][region])
```

```python
# Derechos agrupados por provincia
gb = df.groupby(u'Provincia')
```

```python
for p in gb.groups.keys():
    
    display(HTML('<center><h2>Top 10 derechos otorgados %s</h2></center>'%p))
    display(ddaa.top_ddaa(gb.get_group(p), n=10))
    ax = gb.get_group(p)[ddaa.short['caudal']].plot(title='Derechos otorgados %s'%p, style='x', figsize=(15, 7.5))
    ddaa.zoom(ax)
    plt.show()
```

```python
ddaa.busca(df, 'FISCO')
print 'Derechos reservados por el FISCO en la Region %s:' % region
print '%s [l/s]' % ddaa.busca(df, 'FISCO')[ddaa.short['caudal']].sum()
```

```python
# Agrupa datos por año solicitud
grp_year = df.groupby(lambda x: x.year)
# Sumatoria caudales por año
caudales_year = grp_year[ddaa.short['caudal']].sum()
# Gráfico de barra suma caudales de derechos otorgados por año
caudales_year.plot(kind='bar', figsize=(15, 15), title=u'Caudales Anual Promedio de derechos otorgados Region %s'%region)
```

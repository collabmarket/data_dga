```python
def zoom(ax):
    yi, yf = ax.get_ylim()
    xi, xf = ax.get_xlim()
    mx = (xf + xi) / xf
    my = (yf + yi) / yf
    dx = (xf - xi)
    dy = (yf - yi)
    xi -= xi * (mx / dx)
    xf += xf * (mx / dx)
    yi -= yi * (my / dy)
    yf += yf * (my / dy)
    ax.set_xlim([xi,xf])
    ax.set_ylim([yi,yf])
```
